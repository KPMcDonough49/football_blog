```python
def get_weeks(week_1_games):
    games = dict()
    for i in range(0,len(week_1_games)):
        games[i] = dict()
        games[i]['date'] = week_1_games[i]['scheduled']
        games[i]['game_id'] = week_1_games[i]['id']
        games[i]['home_team'] = week_1_games[i]['home']['name']
        games[i]['away_team'] = week_1_games[i]['away']['name']
        games[i]['home_score'] = week_1_games[i]['scoring']['home_points']
        games[i]['away_score'] = week_1_games[i]['scoring']['away_points'] 
    df = pd.DataFrame.from_dict(games, orient='index')
    return df
```

```python
df = pd.DataFrame()
for i in range (0,len(data['weeks'])):
    week_data = data['weeks'][i]['games']
    df_new = get_weeks(week_data)
    df = df.append(df_new)
```
